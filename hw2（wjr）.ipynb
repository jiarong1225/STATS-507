Your function should ignore spaces and capitalization in assessing whether or not a string is a palindrome, so that "tacocat" and "T A C O cat" are both considered palindromes.

````python
def is_palindrome(string):
  upper_string=string.upper().replace(" ","")
  for i in range(0,len(upper_string)-1):
    if upper_string[i]==upper_string[len(upper_string)-i-1]:
      continue
    else:
      return False
  return True
````

3. <b>(2 points)</b> Write a function called `binary_search` that takes two arguments, a list of integers `t`(which is guaranteed to be sorted in ascending order) and an integer `elmt`, and returns True if `elmt` appears in list `t` and False otherwise. The binary search algorithm works as follows: 

    Search a sorted array by repeatedly dividing the search interval in half. Begin with an interval covering the whole array. If the value of the search key is less than the item in the middle of the interval, narrow the interval to the lower half. Otherwise narrow it to the upper half. Repeatedly check until the value is found or the interval is empty. (Source: https://www.geeksforgeeks.org/binary-search/)
    
    __Note:__ No credit will be given if binary search is not implemented. A simple for loop checking that the element is in the list will not qualify as a binary search.

````python
def binary_search(t, elmt):
  search_key=elmt
  left=0
  right=len(t)-1
  elements_in_interval=right-left+1
  middle=len(t)//2
  while elements_in_interval>1:
    elements_in_interval=right-left+1
    middle=(right+left)//2
    if search_key==t[middle]:
      return True
    elif search_key>t[middle]:
      left+=(elements_in_interval)//2
    elif search_key<t[middle]:
      right-=(elements_in_interval)//2
  return False
````

````python
assert binary_search ([1, 3, 5, 7, 9], 9) == True
assert binary_search ([1, 3, 5, 7, 9], 1) == True
assert binary_search ([1, 3, 5, 7, 9], 7) == True
assert binary_search ([1, 3, 5, 7, 9], 0) == False
````

````python
# YOUR CODE HERE
def check_valid_mx(matrix):
    if not isinstance(matrix,tuple):
        return False
    if len(matrix)==0:
        return False
    length=len(matrix[0])
    for row in matrix:
        if not isinstance(row,tuple):
            return False
        if len(row)!=length:
            return False
        for element in row:
            if not isinstance(element,(int,float)):
                return False
    return True
````

Note: your function should not assume that the input is a dictionary. A valid sparse vector will be a dictionary that has the properties that

    2. all its values are floats. 



````python
# YOUR CODE HERE
def is_valid_sparse_vector(sparse_vector):
    if not isinstance(sparse_vector, dict):
        return False
    for indice,value in sparse_vector.items():
        if not isinstance(indice,int):
            return False
           # raise TypeError("Elements in indices must be integer.")
        if not isinstance(value,float):
            return False
           # raise TypeError("Elements in indices must be float.")
        if indice<=0 or value==0:
            return False
           # raise ValueError("Elements in indices and values cannot be zero.")
    return True
````

````python
# YOUR CODE HERE
def sparse_inner_product(sparse_vecter1,sparse_vecter2):
    if not is_valid_sparse_vector(sparse_vecter1):
        return False
    if not is_valid_sparse_vector(sparse_vecter2):
        return False
    inner_product=0.0
    for index1,value1 in sparse_vecter1.items():
        if index1 in sparse_vecter2:
            value2=sparse_vecter2[index1]
            inner_product+=value1*value2
    return inner_product
````
